# Data cleaning

```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 

df1 = pd.read_csv('steps-1.csv',delimiter = ";")
df1['Mean steps per participant']= df1.mean(axis=1)
df1.head()
```

```python
df2 = pd.read_csv('survey.csv')
df2['weight'].replace({700:0}, inplace=True)
df2 = df2.dropna()
df2.head()
```

```python
df2.describe()
```

```python
df = pd.merge(df1, df2, on = 'id')
df.head()
```

# Exploratory data analysis

### Graph of the distribution 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```python
df['Mean steps per participant'].name = 'Mean steps per participant'
Mean_steps = df['Mean steps per participant'].dropna()
sns.distplot(Mean_steps, kde=False) 
plt.title('Histogram of mean steps per participant')
plt.show()
```

```python
mean_steps = df['Mean steps per participant'].mean()
median_steps = df['Mean steps per participant'].median()

print(f'The median filesize is: {median_steps}')
print(f'The median filesize is: {mean_steps}')

sns.distplot(Mean_steps, kde=False)
plt.title('Histogram of mean steps per participant')
plt.axvline(median_steps, 0, 100, color='green', label='median') 
plt.axvline(mean_steps, 0, 100, color='red', label='mean') 
plt.legend()
plt.show()
```

###  The relation of mean steps per participant with a categorical variable with two levels 

```python
sns.set_theme(style="whitegrid")
gender = df["gender"]
sns.violinplot(x= df["gender"] ,y= df['Mean steps per participant'],hue=df["gender"],
                     palette="muted", split=True)

plt.title('Gender and mean steps per participant')
plt.ylabel('Gender')
plt.xlabel('Mean steps per participant')

plt.show()
```

```python
sns.set(style="darkgrid")
df["age"] = sorted(df["age"])
sns.boxplot( x=df["Mean steps per participant"], y=df["age"],width=0.5)
plt.show()
```

### The relation between two categorical variables

```python
sns.set(style= "darkgrid")
sns.barplot(x=df["age"], y=df["Mean steps per participant"], hue=df["age"],linewidth=0.01)
plt.show()
```

```python
sns.jointplot(x=df["weight"], y=df["Mean steps per participant"], kind="hex", color="#4CB391")
plt.show()
```
